## ðŸ“Œ Context (for report or presentation)

This dataset contains transaction-level data from a coffee shop, including details such as date, time, product purchased, payment type, and revenue. By performing Exploratory Data Analysis (EDA), we aim to uncover customer behavior patterns, sales trends, and business opportunities.

Understanding when and how customers purchase coffee helps answer key business questions such as:

* Which days of the week generate the most revenue?

* What are the peak business hours, and how can staffing be optimized?

* Do customers who pay by cash vs. card spend differently?

* Which coffee products are the top sellers?

* Are there seasonal or monthly trends (e.g., hot drinks in winter, iced drinks in summer)?

These insights provide valuable guidance for:

* Menu planning (e.g., stocking popular items),

* Operational efficiency (e.g., scheduling staff during peak times),

* Marketing strategies (e.g., targeted offers during slow periods),

* Improving customer experience by understanding purchase habits.

## ðŸ”¹ Sales & Revenue Analysis

```python
## 1) What is the total revenue per month, and which month has the highest sales?


highest_sales = data.groupby("Month_name")["money"].sum().reset_index()
print(highest_sales)

plt.figure(figsize=(10,5))

sns.lineplot(x="Month_name", y="money", data=highest_sales, marker="o",palette="Colormap")

plt.title("Monthly Sales Trend", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
```

```python
data.groupby("Month_name")["money"].sum().sort_values(ascending=False).head(1)
```

```python
## 2) Which day of the week generates the most revenue? (e.g., Friday vs Sunday)

print(data.groupby("Weekday")["money"].sum().sort_values(ascending=False).reset_index().head(1))

most_revenue = data.groupby("Weekday")["money"].sum().sort_values(ascending=False).reset_index()


plt.figure(figsize=(8,5))
sns.barplot(x="Weekday",y="money",data=most_revenue,palette="rainbow")
plt.title("Revenue by Weekday", fontsize=14)
plt.xlabel("Day of Week")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
```

```python
## 3) What is the average transaction value per customer?

print(data.groupby("cash_type")["money"].mean())

avg_spending = data.groupby("cash_type")["money"].mean()

plt.figure(figsize=(6,6))
plt.pie(
    avg_spending, 
    labels=avg_spending.index, 
    autopct='%1.1f%%', 
    startangle=90
)

plt.title("Average Spending by Payment Type", fontsize=14)
```

## ðŸ”¹ Product Performance

```python
## 1) Which coffee type sells the most overall?

print(data["coffee_name"].value_counts().sort_values(ascending=False).head(1))

sell = data["coffee_name"].value_counts().reset_index().sort_values(by="coffee_name",ascending=False).head(5)
sell.columns = ["coffee_name","counts"]
sell
```

```python
## 2) Which coffee type generates the highest revenue?

print(data.groupby("coffee_name")["money"].sum().sort_values(ascending=False).head(1))


top_coffees = (
    data.groupby("coffee_name")["money"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

plt.figure(figsize=(8,5))
sns.barplot(x="coffee_name", y="money", data=top_coffees, palette="viridis")

plt.title("Top 5 Coffees by Revenue", fontsize=14)
plt.xlabel("Coffee Name")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
## Is there a seasonal trend (e.g., Hot Chocolate in winter, Iced Latte in summer)?

seasonal_sales = data.groupby(["Month_name", "coffee_name"])["money"].sum().reset_index()

pivot_sales = seasonal_sales.pivot(index="Month_name", columns="coffee_name", values="money").fillna(0)
print(pivot_sales)



pivot_sales.plot(kind="line", figsize=(12,6), marker="o")
plt.title("Seasonal Trend in Coffee Sales", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend(title="Coffee Type")
plt.show()
```

## ðŸ”¹ Customer & Payment Insights

```python
## 1) What percentage of payments are made by card vs cash ?

# Count transactions by payment type

payment_counts = data["cash_type"].value_counts(normalize=True) * 100.
print(payment_counts)

sns.countplot(data=data,x="cash_type",palette="rainbow",stat="percent")
```

```python
## Do cash vs card customers spend differently (average bill size)?

Avg_spending = data.groupby("cash_type")["money"].mean().reset_index()

sns.barplot(data=Avg_spending,x="cash_type",y="money",palette="Set2")
plt.title("Average Bill Size by Payment Type", fontsize=14)
plt.xlabel("Payment Method")
plt.ylabel("Average Spending")
```

```python
## At what time of the day (Morning/Afternoon/Night) do customers spend the most ?

Spend_time = data["Time_of_Day"].value_counts().reset_index()
print(Spend_time)
Spend_time.columns = ["Time_of_Day", "Count"]


sns.barplot(data=Spend_time,x="Time_of_Day",y="Count",palette="pastel")

plt.title("Customer Spending by Time of Day", fontsize=14)
plt.xlabel("Time of Day")
plt.ylabel("Number of Transactions")
```

## ðŸ”¹ Operational Optimization

```python
## Which hour of the day is peak business time? (helps with staffing)

print(data.groupby("Weekday")["hour_of_day"].sum().reset_index().sort_values(by="hour_of_day",ascending=False).head(1))

heatmap_data = (
    data.groupby(["Weekday", "hour_of_day"])["money"]
    .sum()
    .reset_index()
)

pivot_table = heatmap_data.pivot(index="Weekday", columns="hour_of_day", values="money")

# Plot heatmap
plt.figure(figsize=(12,6))
sns.heatmap(pivot_table, cmap="YlGnBu", annot=False, cbar=True)

plt.title("Revenue Heatmap: Weekday vs Hour of Day", fontsize=14)
plt.xlabel("Hour of Day")
plt.ylabel("Weekday")
```

```python
## What is the slowest day/time (potential discount offers) ?

print(data.groupby("Weekday")["hour_of_day"].sum().reset_index().sort_values(by="hour_of_day",ascending=True).head(1))

weekday_revenue = (
    data.groupby("Weekday")["money"]
    .sum()
    .reset_index()
    .sort_values(by="money", ascending=True)
)


plt.figure(figsize=(8,5))
sns.barplot(x="Weekday", y="money", data=weekday_revenue, palette="Blues_r")

plt.title("Revenue by Weekday", fontsize=14)
plt.xlabel("Weekday")
plt.ylabel("Total Revenue")
```

```python
## 
# Define weekend
data["is_weekend"] = data["Weekday"].isin(["Saturday", "Sunday"])


# Average revenue per transaction
avg_revenue = data.groupby("is_weekend")["money"].mean().reset_index()
print(avg_revenue)

```
